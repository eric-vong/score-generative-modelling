```python
import torch
import matplotlib.pyplot as plt

device = "cpu"

loss_5 = torch.load("loss/loss_5_16_32_64_128.pth", map_location=device)
loss_10 = torch.load("loss/loss_10_16_32_64_128.pth", map_location=device)[40:]
# loss_25 = torch.load("loss/loss_list_25_32_64_128_128.pth", map_location=device)
loss_50 = torch.load("loss/loss_50_16_32_64_128.pth", map_location=device)[80:]
loss_100 = torch.load("loss/loss_100_16_32_64_128.pth", map_location=device)[120:]
```

```python
plt.figure(figsize=(10, 10))
labels = [["σ=5", "σ=10"], ["σ=50", "σ=100"]]
loss = [[loss_5, loss_10], [loss_50, loss_100]]

f, axs = plt.subplots(2, 2)

# ---- loop over axes ----
for i, ax in enumerate(axs):
    for j, ax in enumerate(axs):
          axs[i][j].plot(loss[i][j] , label=labels[i][j])
          axs[i][j].legend(loc="upper right")

plt.show()
```

```python
# plt.figure(figsize=())
plt.plot(loss_5, label="σ=5")
plt.plot(loss_10, label="σ=10")
plt.plot(loss_50, label="σ=50")
plt.plot(loss_100, label="σ=100")
plt.legend()
```
